## Ejercicio 1
Escribe una función que convierta números del 1 al 7 en nombres de los dias de la semana. La función constará de un único argumento numérico y una salida de tipo string

````python
def dia_semana(ndia:int)->str:
    match ndia:
        case 1:
            return 'Lunes'
        case 2:
            return 'Martes'
        case 3:
            return 'Miércoles'
        case 4:
            return 'Jueves'
        case 5:
            return 'Viernes'
        case 6:
            return 'Sábado'
        case 7:
            return 'Domingo'
        case _:
            return 'Nono, solo tenemos 7 días de la semana (numérico del 1 al 7)'
````

````python
lista_dias = [1, 2, 3, 4, 5, 6, 7, 8, 0, 'a']

for i in lista_dias:
    print(dia_semana(i))
````

En el ejercicio 8 de flujos de control, creábamos una pirámide invertida de números desde el 5. Crea una función que replique el comportamiento de la pirámide, y utiliza un único parámetro de entrada de la función para determinar el número de filas de la pirámide.

````python
def piramide(n):
    for inicio in range(n, 0, -1):
        print(*range(inicio, 0, -1))  
````

## Ejercicio 3
Escribe una función que compare dos números. La función tiene dos argumentos y hay tres salidas posibles: que sean iguales, que el primero sea  mayor que el segundo, o que el segundo sea mayor que el primero

````python
def comparar(a, b):
    if a < b:
        return f'{a} es menor que {b}'
    if a > b:
        return f'{a} es mayor que {b}'
    return 'Son iguales'
    
````

````python
print(comparar(523576293.0813641, 3547235219.18274))
print(comparar(8.4, 8.4))
````

## Ejercicio 5
Escribe una función que tenga un único argumento, un string. La salida de la función tiene que ser un diccionario con el conteo de todas las letras de ese string, siendo la clave la letra y el valor el conteo.

````python
def cont_dic(texto:str)->dict:
    dic_letras = {}
    for elem in texto:
        dic_letras[elem] = texto.count(elem)
    return dic_letras
````

````python
text = 'Escribe una función que tenga un único argumento, un string'
print(cont_dic(text).items())
````

## Ejercicio 6
Escribir una función que añada o elimine elementos en una lista. La función necesita los siguientes argumentos:
* lista: la lista donde se añadirán o eliminarán los elementos
* comando: "add" o "remove"
* elemento: Por defecto es None.

Tendrá como salida la lista

````python
def add_rmv_list(lista, operacion, item = None):
    if item == None:
        print('No sé qué elemento debo añadir o eliminar')
        return
    else:
        if operacion.lower() == 'add':
            lista.append(item)
            return lista
        elif operacion.lower() == 'remove':
            lista.remove(item)
            return lista
        else:
            print('No reconozco el comando')
            return
        
````

````python
ejer_6 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
print(ejer_6)

add_rmv_list(ejer_6, 'add', 27)
print(ejer_6)

add_rmv_list(ejer_6, 'remove', 90)
print(ejer_6)

add_rmv_list(ejer_6, 'count', 27)
print(ejer_6)

add_rmv_list(ejer_6, 'remove')
print(ejer_6)
````

## Ejercicio 7
Crea una función que reciba un número arbitrario de palabras, y devuelva una frase completa, separando las palabras con espacios.

````python
def frase(*args):
    for elem in args:
        frase = ' '.join(elem)
    return frase
````

````python
palabras = ['Crea', 'una', 'función', 'que', 'reciba', 'un', 'número', 'arbitrario', 'de', 'palabras']
print(palabras)
print(frase(palabras))
````

## Ejercicio 8
Escribe un programa que obtenga el enésimo número de la [serie de Fibonacci](https://es.wikipedia.org/wiki/Sucesi%C3%B3n_de_Fibonacci). Tienes que crear una función recursiva con un único argumento.

```Python
fibonacci(2) = 1
fibonacci(4) = 3
fibonacci(5) = 5
fibonacci(8) = 21
```

fibonacci(n) = fibonacci(n-1) + fibonacci(n-2)

````python
def fibonacci(n):
    if n == 1:
        return 1
    elif n == 2:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)
````

````python
print(fibonacci(1))
print(fibonacci(2))
print(fibonacci(3))
print(fibonacci(4))
print(fibonacci(5))
print(fibonacci(6))
print(fibonacci(7))
print(fibonacci(8))
````

## Ejercicio 9
Define en una única celda las siguientes funciones:
* Función que calcule el área de un cuadrado
* Función que calcule el área de un triángulo
* Función que calcule el área de un círculo

En otra celda, calcular el area de:
* Dos círculos de radio 10 + un triángulo de base 3 y altura 7
* Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4

````python
import math

def a_cuad(lado):
    return 4 * lado

def a_tri(base, altura):
    return base * altura / 2

def a_circ(radio):
    return math.pi * radio**2
````

````python
# Dos círculos de radio 10 + un triángulo de base 3 y altura 7
r = 10
b = 3
h = 7

print(round(2*a_circ(r) + a_tri(b, h),3))

# Un cuadrado de lado = 10 + 3 círculos (uno de radio = 4 y los otros dos de radio = 6) + 5 triángulos de base = 2 + altura = 4
l = 10
r1 = 4
r2 = 6
b = 2
h = 4

print(round(a_cuad(l) + a_circ(r1) + 2*a_circ(r2) + 5*a_tri(b,h),3))
````
