## Ejercicio 2
Escribe un programa para pasar de grados a radianes. Hay que usar `input`. Recuerda que la conversión se realiza mediante

$$radianes = grados*(pi/180)$$

````python
import math
grados = round(float(input("Ingrese un ángulo en grados: ")), 2)
radianes = round(grados * math.pi / 180, 2)
print(f"{grados} grados son {radianes} radianes.")
````

## Ejercicio 3
Escribe un programa que calcule el area de un paralelogramo (base x altura). También con `input`

````python
base = round(float(input("Ingrese la base del paralelogramo: ")), 2)
altura = round(float(input("Ingrese la altura del paralelogramo: ")), 2)
area = round(base * altura, 2)
print(f"El área del paralelogramo es: {area}")
````

## Ejercicio 6
Mismo programa que en 5, pero en esta ocasión tienen que ser tres inputs y dos salidas. Una de las salidas que nos indique si todos son iguales, y la otra si al menos dos inputs sí que lo son

````python
input1 = input("Introduce primer valor: ")
input2 = input("Introduce segundo valor: ")
input3 = input("Introduce tercer valor: ")

print('¿Son todos iguales?', input1 == input2 == input3)
print('¿Hay al menos dos iguales?', input1 == input2 or input1 == input3 or input2 == input3)
````

## Ejercicio 7
Crea un programa que recoja dos inputs. Tiene que comprobar si su suma es igual, superior o inferior a 10, cada condición por separado

````python
input1 = input("Introduce primer valor: ")
input2 = input("Introduce segundo valor: ")
suma = int(input1) + int(input2)

print("Los valores suman 10", suma == 10)
print("La suma es mayor que 10", suma > 10)
print("La suma es menor que 10", suma < 10)
````
