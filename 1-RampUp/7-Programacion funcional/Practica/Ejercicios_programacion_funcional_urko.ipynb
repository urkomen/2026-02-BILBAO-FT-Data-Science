### 1. Utiliza map para elevar al cuadrado cada número en la lista [1, 2, 3, 4, 5].

Solución esperada: [1, 4, 9, 16, 25]

```python
lista_1 = [1, 2, 3, 4, 5]

lista_cuadr = list(map(lambda x: x**2, lista_1))
print(list(lista_cuadr))
```

### 2. Filtrar números pares de una lista:

Emplea filter para obtener sólo los números pares de la lista [1, 2, 3, 4, 5, 6].

Solución esperada: [2, 4, 6]

```python
lista_2 = [1, 2, 3, 4, 5, 6]
fil_2 = filter(lambda x: x%2 == 0, lista_2)

print(list(fil_2))
```

### 3. Convertir grados Celsius a Fahrenheit:

Usa map y una función lambda para convertir una lista de temperaturas en grados Celsius [0, 10, 20, 30] a Fahrenheit.

Solución esperada: [32.0, 50.0, 68.0, 86.0]

```python
celsius = [0, 10, 20, 30]

Farenheit = list(map(lambda x: x*9/5 + 32, celsius))

print(Farenheit)
```

### 4. Sumar los cuadrados de los números pares de una lista:

Combina filter, map y reduce (importa reduce desde functools) para sumar los cuadrados de los números pares en [1, 2, 3, 4, 5, 6].

Solución esperada: 56

```python
from functools import reduce

lista_4 = [1, 2, 3, 4, 5, 6]

pares = list(filter(lambda x: x%2 == 0, lista_4))

cuadr = list(map(lambda x: x**2, pares))

suma_cuadr_pares = reduce(lambda x,y: x+y, cuadr)

print(suma_cuadr_pares)
```

```python
# Al estilo Unai
from functools import reduce

lista_4_2 = [1, 2, 3, 4, 5, 6]

print(reduce(lambda x,y: x+y, list(map(lambda x: x**2, list(filter(lambda x: x%2 == 0, lista_4_2))))))
```

### 5. Encadenar map, filter, y reduce para resolver un problema complejo:

Dada una lista de strings que representan números enteros y "NA" (por ejemplo, ["2", "5", "NA", "9", "10"]), convierte los números a enteros, filtra los "NA", suma el cuadrado de los números que son mayores que 5 y menores que 10.

Solución esperada: Un único valor que es la suma de los cuadrados de los números válidos.

```python
from functools import reduce

lista_5 = ["2", "5", "NA", "9", "10"]

lista_5_sinNA = list(filter(lambda x: x.isdigit(), lista_5))

num_int = list(map(lambda x: int(x), lista_5_sinNA))

int_mayor5 = list(filter(lambda x: 5 < x < 10, num_int))

cuadr = list(map(lambda x: x**2, int_mayor5))

suma_cuadr_pares = reduce(lambda x,y: x+y, cuadr)

print(suma_cuadr_pares)
```

```python
# Otra manera más reducida juntando pasos
from functools import reduce

lista_5_2 = ["2", "5", "NA", "9", "10"]

lista_5_sinNA = list(filter(lambda x: x.isdigit(), lista_5_2))

# num_int = list(map(lambda x: int(x), lista_5_sinNA))

# int_mayor5 = list(filter(lambda x: 5 < x < 10, num_int))

cuadr = list(map(lambda x: (int(x))**2 if 5 < int(x) < 10 else 0, lista_5_sinNA))

suma_cuadr_pares = reduce(lambda x,y: x+y, cuadr)

print(suma_cuadr_pares)
```

```python
# Al estilo Unai
from functools import reduce

lista_5_3 = ["2", "5", "NA", "9", "10"]

suma_cuadr_pares = reduce(lambda x,y: x+y, list(map(lambda x: (int(x))**2 if 5 < int(x) < 10 else 0, list(filter(lambda x: x.isdigit(), lista_5_3)))))

print(suma_cuadr_pares)
```
