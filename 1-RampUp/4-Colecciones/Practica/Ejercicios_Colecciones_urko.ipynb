## Ejercicio 3
Crea una lista nueva con todas las combinaciones de las siguientes dos listas:
> ```ejer_3_1 = ["Hola", "amigo"]```
>
> ```ejer_3_2 = ["Que", "tal"]```

Obten el siguiente output:

```['Hola Que', 'Hola tal', 'amigo Que', 'amigo tal']```

````python
ejer_3_1 = ["Hola", "amigo"]
ejer_3_2 = ["Que", "tal"]
````

````python
combos = [] # Combinaciones

for i in ejer_3_1:
    for j in ejer_3_2:
        combos.append(i + ' ' + j)
print(combos)
````

````python
# Ejercicio resuelto por David
ejer_3_1 = ['1', '2'] # Modifico las listas para comprobar efecto de repeat en el método itertools.product
ejer_3_2 = ['3', '4']

import itertools
ejer_3_results = []

for combination in itertools.product(ejer_3_1, ejer_3_2, repeat=3):
    print(combination)
    string = ""
    for value in combination:
        string = string + " " + value 
    ejer_3_results.append(string[1:])

print(ejer_3_results)
````

## Ejercicio 4
Dada la siguiente lista, encuentra por el índice el valor 45 y sustitúyelo por el 0

## Ejercicio 5
Dada la siguiente lista, elimina todos los valores iguales a 3
> ```ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]```

TIP: No intentes eliminar elementos sobre la lista que estás iterando. Haz una copia con `ejer_5.copy()`.

````python
ejer_5 = [3, 20, 3, 47, 19, 3, 29, 45, 67, 78, 90, 3, 3]
````

````python
ejer5_copy = ejer_5.copy()
print(ejer5_copy)

while 3 in ejer5_copy:
    ejer5_copy.remove(3)

print(ejer5_copy)
````

````python
# Ejercicio resuelto pod David
for value in ejer_5.copy():
    if value == 3:
        ejer_5.remove(value)

print(ejer_5)

# Otra manera
while 3 in ejer_5:
    ejer_5.remove(3)

print(ejer_5)

# Otra manera
ejer_5_results = [x for x in ejer_5 if x != 3]
print(ejer_5_results)
````
