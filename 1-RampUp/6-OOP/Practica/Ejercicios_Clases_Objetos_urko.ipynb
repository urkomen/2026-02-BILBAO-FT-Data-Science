## Ejercicio 1
Somos una empresa que tiene varias tiendas de electrodomésticos. Necesitamos un programa para manejar las tiendas, ver las ventas que han tenido, cuántos empleados hay, etc... Para ello vamos a modelizar la tienda en una clase, que tendrá las siguientes características:
* Nombre clase: "Tienda"
* Atributos comunes:
    * Tipo: "Electrodomésticos"
    * Abierta: True
* Atributos propios de cada tienda:
    * Nombre - String
    * Dirección - String
    * Número de empleados - int
    * Ventas últimos 3 meses - Lista de 3 numeros
* Método para calcular las ventas de todos los meses, que devuelva un número con todas las ventas.
* Método que calcula la media de ventas de los últimos meses, por empleado
* Método que devuelve en un string el nombre de la tienda, junto con su dirección.
* Método que obtiene las ventas del último mes.
* Método para dar una proyección de las ventas en caso de haber invertido X dinero en marketing en esa tienda. Siendo X el parámetro de entrada. Si X es menor de 1000, las ventas de los ultimos 3 meses hubiesen sido de (1.2 x ventas), si es mayor o igual de 1000, las ventas hubiesen sido de (1.5 x venta). El método reescribe el atributo de ventas en los últimos 3 meses, y además devuelve las ventas con el aumento.

Se pide:
1. Implementar la clase "Tienda"
2. Crear tres tiendas con datos inventados
3. Comprueba en al menos una de ellas, todo lo implementado en la clase tienda (sus atributos, media de ventas, ventas/empleado...)
3. Calcular las ventas del último mes de todas las tiendas. Para ello usa el bucle `for`
4. Imprime por pantalla los nombres de las tiendas cuya dirección lleve el string "Avenida"
5. Seguro que ya has hecho este apartado... Documenta la clase :)

```python
# 1
class Tienda():    
    def __init__(self, nom_tienda, direc_tienda, num_empleados, ventas3meses):
        self.nombre = nom_tienda
        self.direccion = direc_tienda
        self.empleados = num_empleados
        self.ventas3 = ventas3meses
        self.tipo = 'Electrodomésticos'
        self.abierta = True

    # Calcula la suma de todas las ventas los últimos 3 meses
    def total_ventas(self):
        return sum(self.ventas3)
        
    # Calcula la media de ventas de los últimos 3 meses por empleado
    def media_ventas(self):
        return round(self.total_ventas() / self.empleados, 3) # Redondeamos para que no se muestren tantos decimales

    # Nombre de la tienda
    def nombre_tienda(self):
        return self.nombre
    
    # Ventas del último mes
    def ventas_ult_mes(self):
        return self.ventas3[-1]

    # Proyección de ventas. Los valores de las ventas se cambian
    def proyec_invers(self, X):
        if X < 1000:
            self.ventas3 = [1.2 * i for i in self.ventas3]
        elif X >= 1000:
            self.ventas3 = [1.5 * i for i in self.ventas3]
        return self.ventas3
```

```python
# 2
media = Tienda("MediaMarkt", "Avenida de la gitana 7", 25, [1000, 1200, 1500])
fnac = Tienda("Fnac", "Calle Librerías 12", 327, [12000, 13450, 18330])
yao = Tienda("Yao Ming - Electrodomésticos Made In China", "Avenida San Vicente 3", 89, [2100, 2300, 2200])

# 3 Comprobamos atributos y métodos con una de las tiendas
print('Los atributos son')
print('Nombre:', fnac.nombre)
print('Dirección:', fnac.direccion)
print('Número de empleados:', fnac.empleados)
print('Ventas de los últimos 3 meses:', fnac.ventas3)
print('Tipo de tienda:', fnac.tipo)
print('Está abierta?', fnac.abierta)

print('Todas las ventas los últimos 3 meses son', fnac.total_ventas())
print('La media de ventas por empleado es', fnac.media_ventas())
print('Nombre de la tienda:', fnac.nombre_tienda())
print('Ventas del último mes:', fnac.ventas_ult_mes())
inversion = 750
inversion2 = 5580
print(f'Proyección de ventas con una inversión de {inversion}: {fnac.proyec_invers(inversion)}')
print(f'Proyección de ventas con una inversión de {inversion2}: {fnac.proyec_invers(inversion2)}')
```

```python
# 6 Dejamos más "bonito" y documentamos la clase Tienda para lectura fácil
class Tienda:
    '''
    Representa una tienda física con información básica como nombre,
    dirección, número de empleados y ventas trimestrales.
    '''

    def __init__(self, nom_tienda: str, direc_tienda: str, num_empleados: int, ventas3meses: list):
        '''
        Inicializa una nueva tienda.

        Parámetros:
            nom_tienda (str): Nombre de la tienda.
            direc_tienda (str): Dirección física de la tienda.
            num_empleados (int): Número de empleados.
            ventas3meses (list[float] | list[int]): Ventas de los últimos 3 meses.
        '''
        self.nombre = nom_tienda
        self.direccion = direc_tienda
        self.empleados = num_empleados
        self.ventas3 = ventas3meses
        self.tipo = 'Electrodomésticos'
        self.abierta = True

    def total_ventas(self) -> float:
        '''
        Calcula la suma total de ventas de los últimos 3 meses.

        Returns:
            float: Total de ventas.
        '''
        return sum(self.ventas3)

    def media_ventas(self) -> float:
        '''
        Calcula la media de ventas por empleado en los últimos 3 meses.

        Returns:
            float: Media de ventas por empleado.
        '''
        return self.total_ventas() / self.empleados

    def nombre_tienda(self) -> str:
        '''
        Devuelve el nombre de la tienda.

        Returns:
            str: Nombre de la tienda.
        '''
        return self.nombre

    def ventas_ult_mes(self) -> float:
        '''
        Devuelve las ventas del último mes registrado.

        Returns:
            float: Ventas del último mes.
        '''
        return self.ventas3[-1]

    def proyec_invers(self, X: float) -> list:
        '''
        Proyecta las ventas en función de la inversión realizada.
        IMPORTANTE: El valor de ventas del objeto se cambia con 
                    los nuevos valores calculados.

        Parámetros:
            X (float): Cantidad invertida.

        Returns:
            list[float]: Lista de ventas proyectadas.
        '''
        if X < 1000:
            self.ventas3 = [1.2 * i for i in self.ventas3]
        else:
            self.ventas3 = [1.5 * i for i in self.ventas3]

        return self.ventas3
```

## Ejercicio 2
En este ejercicio vamos a implementar una clase *Perro* en Python. La clase tiene las siguientes características:
* Cosas que sabemos seguro que tiene un perro
    * Tiene 4 patas
    * 2 orejas
    * 2 ojos
    * Una velocidad de 0. Por defecto, el perro está parado
* Cuando se inicialice:
    * El perro será de una determinada raza
    * Por defecto tendrá pelo "Marrón", a no ser que se diga lo contrario.
    * Por defecto no tendrá dueño, a no ser que se diga lo contrario.
    
* Dispondrá también de un método llamado andar, que tiene un argumento de entrada (aumento_velocidad). Este valor se le sumará a la velocidad que ya llevaba el perro.
* Necesita otro método (parar), donde pondremos la velocidad a 0.
* Otro método llamado "ladrar", que tendrá un argumento de entrada, y la salida será el siguiente string: "GUAU!" + el argumento de entrada.


Se pide:
* Implementa la clase *Perro*
* Crea un objeto de tipo *Perro*, sin dueño
* Comprueba que están bien todos sus atributos
* Prueba que ande, y comprueba su velocidad
* Páralo
* Documenta la clase *Perro*

```python
# 1
class Perro():    
    # Atributos comunes
    patas = 4
    orejas = 2
    ojos = 2
    velocidad = 0
    
    def __init__(self, raza_perro, pelo_perro = 'Marrón', owner_perro = False):
        self.raza = raza_perro
        self.pelo = pelo_perro
        self.owner = owner_perro

    # Aumenta velocidad del perro
    def andar(self, aumento_vel) -> float:
        self.velocidad += aumento_vel
        return self.velocidad
        
    # Calcula la media de ventas de los últimos 3 meses por empleado
    def parar(self):
        self.velocidad = 0
        return self.velocidad

    # Nombre de la tienda
    def ladrar(self, ladrido):
        return 'GUAU! ' + ladrido
```

```python
# 5 Dejamos más "bonito" y documentamos la clase Perro para lectura fácil
class Perro:
    '''
    Representa un perro con atributos básicos como raza, velocidad,
    color del pelo y si tiene dueño.
    '''

    def __init__(self, raza_perro: str, vel_perro: float = 0,
                 pelo_perro: str = 'Marrón', owner_perro: bool = False):
        '''
        Inicializa un nuevo perro.

        Parámetros:
            raza_perro (str): Raza del perro.
            vel_perro (float): Velocidad inicial del perro (por defecto 0).
            pelo_perro (str): Color del pelo del perro.
            owner_perro (bool): Indica si el perro tiene dueño.
        '''
        self.raza = raza_perro
        self.velocidad = vel_perro
        self.pelo = pelo_perro
        self.owner = owner_perro

        # Atributos físicos fijos
        self.patas = 4
        self.orejas = 2
        self.ojos = 2

    def andar(self, aumento_vel: float) -> float:
        '''
        Aumenta la velocidad del perro.

        Parámetros:
            aumento_vel (float): Cantidad en la que aumenta la velocidad.

        Returns:
            float: Nueva velocidad del perro.
        '''
        self.velocidad += aumento_vel
        return self.velocidad

    def parar(self) -> float:
        '''
        Detiene al perro.

        Returns:
            float: Velocidad actual (0).
        '''
        self.velocidad = 0
        return self.velocidad

    def ladrar(self, ladrido: str) -> str:
        '''
        Hace que el perro ladre.

        Parámetros:
            ladrido (str): Texto adicional al ladrido.

        Returns:
            str: Texto de ladrido y el mensaje adicional.
        '''
        return 'GUAU!' + ladrido
```
