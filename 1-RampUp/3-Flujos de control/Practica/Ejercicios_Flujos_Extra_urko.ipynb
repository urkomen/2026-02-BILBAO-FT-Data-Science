## Ejercicio Extra
Programa una calculadora que

* Pregunte qué operación quieres realizar (multiplicación/división/suma/resta/potencia/salir)
* Pida dos números 
* Realice la operación
* Siga funcionando hasta que el usuario elija salir

#### Bonus:
* Que guarde todas las operaciones realizadas y las imprima por pantalla cuando el usuario elija esa opción
* Que sea robusta frente a entradas erróneas del usuario o frente a operaciones como dividir por 0

```python
import math
import pprint
import time
```

```python
# Tenemos en cuenta mayúsculas y tildes que pueda escribir el usuario
operacion = input('Introduce la operación a realizar (multiplicación/división/suma/resta/potencia/historial/salir): ').lower().replace('ó','o') 
hist_oper = []

while operacion != 'salir':
    if operacion != 'historial' or operacion != 'hist':
        print('Muy bien, necesito dos números con los que operar.')
        num1 = float(input('Introduce el primer número').replace(',','.'))
        print(f'Primer número: {num1}')
        num2 = float(input('Introduce el segundo número').replace(',','.'))
        print(f'Segundo número: {num2}')
    
    match operacion:
        case 'suma' | '+':
            operador = '+' # Guardamos símbolos para mostrar y recoger en el historial
            result = num1 + num2
        case 'resta' | '-':
            operador = '-'
            result = num1 - num2
        case 'multiplicacion' | '*':
            operador = '*'
            result = num1 * num2
        case 'division' | '/':
            if num2 != 0: # Asegurarse de que num2 no sea 0
                operador = '/'
                result = num1 / num2
            else:
                print('Hey, canalla! No puedo dividir por 0')
                continue # Volvemos a intentarlo
        case 'potencia' | '**':
            operador = '**'
            result = num1 ** num2
            
        case 'historial' | 'hist':
            print('Imprimimos historial de operaciones')
            for i, oper in enumerate(hist_oper):
                print(f'{i}. {oper}')
            
            # Pedimos la siguiente operación al usuario
            operacion = input('Introduce la operación a realizar (multiplicación/división/suma/resta/potencia/historial/salir): ').lower().replace('ó','o')
            continue # Volvemos al inicio del bucle
        case _:
            print('No reconozco la operación')
            operacion = input('Introduce la operación a realizar (multiplicación/división/suma/resta/potencia/historial/salir): ').lower().replace('ó','o')
            continue # Volvemos al inicio del bucle

    hist_oper.append(f'{num1} {operador} {num2} = {result}')
    print(f'{num1} {operador} {num2} = {result}')
    
    # Pedimos la siguiente operación al usuario
    operacion = input('Introduce la operación a realizar (multiplicación/división/suma/resta/potencia/historial/salir): ').lower().replace('ó','o')
    
print('De acuerdo. Hasta la próxima!!')
    
```
