5. Usando una figura Y EJES POR SEPARADO, usa numpy para los valores del eje X entre 0 y 5. Pinta dos gráficas en dos cajas distintas, a la izquierda una recta con pendiente positiva de 3 que pase por (0,0) y a la derecha una recta con pendiente negativa de 3 que pase por (0,-5). Elige la precisión en el eje X que desees.  

```python
# Creamos figura y obtenemos valores de x de 0 a 5
plt.figure()
x = np.linspace(0, 5, 100)

# graf. 1: y = 3x
plt.subplot(1, 2, 1)
plt.plot(x, 3*x, label = 'y = 3*x')
plt.title("recta sube")
plt.xlabel("eje X")
plt.ylabel("eje Y")

# graf. 2: y = -3x-5
plt.subplot(1, 2, 2)
plt.plot(x, -3*x - 5, label = 'y = -3*x - 5')
plt.title("recta baja")
plt.xlabel("eje X")
plt.ylabel("eje Y")

plt.legend();
```

7. Crea un gráfico de barras. Pintamos la altura de un grupo de amigos, cada barra representa a una persona. Ana mide 160 cm, Luis mide 180 cm, Pedro mide 175 cm, Sofía mide 190 cm, Carmen mide 170 cm. Las barras serán verdes.
Consejo: si no vas a tener que manipular especialmente los ejes, es más sencillo dejarlos dentro de la figura.

```python
amigos = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']
altura = [160, 180, 175, 190, 170]

plt.bar(amigos, altura, color = 'green')
plt.title("Altura de amguetes")
plt.xlabel("Amigos")
plt.ylabel("Altura")
```

8. Cambia el tamaño de la figura a (3,5)

```python
plt.figure(figsize=(3, 5))  # ancho=10, alto=5

plt.bar(amigos, altura, color = 'green')
plt.title("Altura de amguetes")
plt.xlabel("Amigos")
plt.ylabel("Altura")
```

9. Basándote en el gráfico anterior, borra la escala del eje Y (ya aparece en la altura)
Pista: cuando plt.yticks() recibe una lista vacía no pinta el eje Y

```python
plt.figure(figsize=(3, 5))  # ancho=10, alto=5

barras = plt.bar(amigos, altura, color = 'green')
# Meto los valores de las barras que vamos a quitar del eje Y
for barra in barras:
    altura = barra.get_height()
    plt.text(barra.get_x() + barra.get_width()/2,
             altura,
             str(altura),
             ha='center',
             va='bottom')


plt.title("Altura de amguetes")
plt.xlabel("Amigos")
plt.ylabel("Altura")

plt.yticks([])
```

10. Prueba a hacer el gráfico con las barras en horizontal.

```python
plt.figure(figsize=(3, 5))  # ancho=10, alto=5
amigos = ['Ana', 'Luis', 'Pedro', 'Sofía', 'Carmen']
altura = [160, 180, 175, 190, 170]

barras = plt.barh(amigos, altura, color = 'green')
# Meto los valores de las barras que vamos a quitar del eje Y
for barra in barras:
    ancho = barra.get_width()
    y = barra.get_y() + barra.get_height()/2
    plt.text(ancho, y, str(ancho), va='center', ha='left')

plt.title("Altura de amguetes")
plt.xlabel("Altura")
plt.ylabel("Amigos")

plt.xticks([])
```
