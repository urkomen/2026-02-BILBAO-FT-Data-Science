````python
import requests
import pandas as pd

# ==========================================================
# CONFIGURACIÓN
# ==========================================================
API_KEY = "TU_API_KEY_AQUÍ"  # El alumno debe poner su clave
BASE_URL = "https://api.themoviedb.org/3"
INICIAL = "M"

# ==========================================================
# 1. OBTENCIÓN DEL MAESTRO DE GÉNEROS
# ==========================================================
def get_genres_dict(api_key):
    """
    Consulta el endpoint de géneros y devuelve un diccionario {id: nombre}
    """
    url = f"{BASE_URL}/genre/movie/list"
    params = {'api_key': api_key, 'language': 'es-ES'}
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        genres_list = response.json()['genres']
        # Convertimos la lista en un diccionario para búsqueda rápida: {28: 'Acción', ...}
        return {g['id']: g['name'] for g in genres_list}
    else:
        print(f"Error al obtener géneros: {response.status_code}")
        return {}

# ==========================================================
# 2. BÚSQUEDA DE PELÍCULAS
# ==========================================================
def search_movies_by_initial(api_key, query):
    """
    Busca películas que coincidan con la inicial
    """
    url = f"{BASE_URL}/search/movie"
    params = {
        'api_key': api_key,
        'query': query,
        'language': 'es-ES',
        'include_adult': 'false'
    }
    
    response = requests.get(url, params=params)
    
    if response.status_code == 200:
        return response.json()['results']
    else:
        print(f"Error en la búsqueda: {response.status_code}")
        return []

# ==========================================================
# 3. PROCESAMIENTO PRINCIPAL
# ==========================================================

# Paso A: Obtener el mapa de géneros
genres_map = get_genres_dict(API_KEY)

# Paso B: Buscar las películas
movies_raw = search_movies_by_initial(API_KEY, INICIAL)

# Paso C: Limpieza y transformación de datos
processed_data = []

for movie in movies_raw:
    # Transformar la lista de IDs de géneros en nombres reales
    # Usamos .get() para evitar errores si un ID no existe en nuestro mapa
    genre_ids = movie.get('genre_ids', [])
    genre_names = [genres_map.get(gid) for gid in genre_ids if genres_map.get(gid)]
    
    # Creamos el diccionario con la estructura requerida
    movie_info = {
        'id': movie.get('id'),
        'title': movie.get('title'),
        'release_date': movie.get('release_date'),
        'genres': ", ".join(genre_names), # Convertimos la lista ['Acción', 'Drama'] en "Acción, Drama"
        'vote_average': movie.get('vote_average'),
        'overview': movie.get('overview')
    }
    processed_data.append(movie_info)

# ==========================================================
# 4. CREACIÓN DEL DATAFRAME Y EXPORTACIÓN
# ==========================================================
df_movies = pd.DataFrame(processed_data)

# Mostramos los primeros resultados
print(f"Se han procesado {len(df_movies)} películas.")
display(df_movies.head())

# Exportar a CSV
df_movies.to_csv("peliculas_tmdb.csv", index=False, encoding='utf-8')
print("Archivo 'peliculas_tmdb.csv' generado con éxito.")
````
