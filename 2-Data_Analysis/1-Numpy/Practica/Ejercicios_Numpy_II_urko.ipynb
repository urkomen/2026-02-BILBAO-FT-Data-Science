## Ejercicio 12
Encuentra los índices de los valores que no sean 0 del siguiente vector [1,2,0,0,4,0]. El output sería [1, 2 4]. [Encuentra el método apropiado en la documentación.](https://numpy.org/doc/) Encapsúlalo en una función, con un argumento de entrada de un vector.

````python
# Yo no he entendido de manera en que devuelva 
# el mismo array sin 0s. 
# He entendido que devuelve un array de índices de
# los elementos diferentes a 0 
def quitar_0s(array):
    return array[array != 0]

def indices_no0s(array):
    return np.where(array != 0)
````

````python
array_12 = np.array([1,2,0,0,4,0])
print(array_12)

# Quito los 0s del array
print(quitar_0s(array_12))

# Devuelvo índices de los elementos que no son 0
array_12 = np.array([1,2,0,0,4,0])
print(indices_no0s(array_12))
````

## Ejercicio 15
Crea un array cuadrado de 2D, con las dimensiones que quieras, pero que todo el borde esté compuesto por 1s, y el interior por 0s

````python
import random
n = int(random.randint(2,6)) # Matrices cuadradas de 2 a 6 dimensiones

array_15 = np.zeros((n,n), dtype = int)
array_15[0, :] = 1
array_15[-1, :] = 1
array_15[:, 0] = 1
array_15[:, -1] = 1
print(array_15)
````

## Ejercicio 18
Crea una matriz aleatoria de 5x5 y normaliza sus valores. Para normalizar tienes que restar sus valores por la media(`np.mean`) y dividir el resultado por la desviación estándar(`np.std`).

````python
np.random.seed(42)
array_18 = np.random.rand(5,5)
print(array_18)

media = array_18.mean()
desv = array_18.std()
norm = (array_18 - media) / desv
print(norm)


````

## Ejercicio 19
1. Crea una secuencia de números que vayan del 0 al 10.
2. Crea un nuevo array, que parta del anterior, pero sustituya por el valor -1, todos los elementos cuyo valor se encuentre entre el 3 y el 8.
3. Crea un nuevo array que niegue todos los elementos entre el 3 y el 8

````python
array_19 = np.arange(11)
print(array_19)

array_19_2 = np.where((array_19 >= 3) & (array_19 <= 8), -1, array_19)
print(array_19_2)

# Negar matemáticamente es buscar el número opuesto, en este caso cambiar el signo a un número
array_19_3 = np.where((array_19 >= 3) & (array_19 <= 8), -array_19, array_19)
print(array_19_3)

print(array_19)
````
