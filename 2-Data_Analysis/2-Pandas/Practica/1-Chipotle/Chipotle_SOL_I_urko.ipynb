### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv) and assign it to a variable called chipo.

```python
url = "https://raw.githubusercontent.com/justmarkham/DAT8/master/data/chipotle.tsv"
chipo = pd.read_csv(url, sep="\t")
chipo
```

```python
chipo.info()
```

### Step 3. Name of the max valued product

```python
# Revisamos los diferentes valores de 'item_price'
chipo['item_price'].value_counts()

# Suponemos que todos los precios tienen la misma estructura
# Creamos nueva columna con el precio sin el signo dólar
chipo["price_float"] = chipo["item_price"].str.replace("$", "").astype(float)
chipo

# Buscamos el valor máximo de 'price_float', obtener su índice y buscar en 'item_name'
prod_caro = chipo.loc[(chipo['price_float']/chipo['quantity']).idxmax(), 'item_name']
print('El producto más caro es', prod_caro)
```

```python
chipo[chipo['item_name'] == 'Steak Salad Bowl']
```

### Step 4. How many products cost more than $10.00?

```python
# Cuento 'item_name' para que no me salgan todas las columnas
# Verifico que en esa columna no hay nulos
print(len(chipo) == chipo['item_name'].count())

# Contar elementos con precio mayor de $10.00
chipo['item_name'][chipo['price_float'] > 10].count()
```

### Step 4.1: Y cuántos pedidos se han hecho con un producto de más de 10$? Es lo mismo?

```python
chipo.head(50)
```

```python
# En un mismo pedido podría haber más de un producto de más de 10$
# Contamos número de pedidos que contengan al menos un producto de más de 10$
num_pedidos = chipo['order_id'][chipo['price_float'] > 10].nunique()

print('Hay', num_pedidos, 'pedidos con al menos un producto de más de 10$')
```

### Step 4.2: Y cuántos pedidos se han hecho de más de 10$? Es lo mismo?

```python
# Para un mismo order_id, la suma de todos los precios debe ser mayor que 10$

# Agrupamos los pedidos, sumando los precios de los productos y los contamos
num_pedidos2 = chipo.groupby('order_id')['price_float'].sum().count()

print('Hay', num_pedidos2, 'pedidos de más de 10$')
```

### Step 4.3: Y en cuántos pedidos se ha pagado más de 10$ por un mismo producto? Es lo mismo?

```python
# Buscar pedidos con productos con el mismo nombre que sumen más de 10$

# Pro pedido, agrupamos los mismos productos y sumamos los precios para cada producto
pedidos3 = (chipo.groupby(['order_id', 'item_name'])['price_float']
            .sum()
            .reset_index())

# Buscamos los pedidos que tengan algún producto que sume más de 10$
num_pedidos_10 = pedidos3['order_id'][pedidos3['price_float'] > 10].count()

print('Hay', num_pedidos_10, 'pedidos donde un mismo un producto sume más de 10$')
```

### Step 5. What is the price of each item and name it unit_price. Get only item_name and unit_price

```python
# Según lo que entiendo, ya lo he hecho en el primer punto
# La última columna 'price_float' equivale a 'unit_price'
chipo['unit_price'] = chipo['price_float'] / chipo['quantity']
chipo
```

```python
chipo[['item_name','unit_price']]
```

### Step 6. Sort by the name of the item

```python
chipo_ord = chipo.sort_values('item_name', ascending=True)
chipo_ord
```

### Step 7. What was the quantity of the most expensive item ordered? 2 ways

```python
# Recuperamos el nombre del producto más caro
producto_caro = chipo.loc[chipo['unit_price'].idxmax(), 'item_name']

# Calcula la suma de la cantidad de ese producto en diferentes pedidos
veces_pedido7 = chipo.loc[chipo['item_name'] == producto_caro, 'quantity'].sum()

print('El producto con el mayor precio unitario se ha vendido', veces_pedido7, 'veces.')
```

```python
# Ordenar la lista de mayor a menor 'unit_price' y nos quedamos con el producto

producto_caro2 = chipo.nlargest(1, "unit_price")["item_name"].iloc[0]
producto_caro2
veces_pedido7_2 = chipo[chipo["item_name"] == producto_caro2]["quantity"].sum()

print('El producto con el mayor precio unitario se ha vendido', veces_pedido7_2, 'veces.')
```

### Step 8. How many times was a Veggie Salad Bowl ordered?

```python
ventas_veggie = chipo.loc[chipo['item_name'] == 'Veggie Salad Bowl']['quantity'].sum()

print('Se han vendido', ventas_veggie, 'Veggie Salad Bowl.')
```

### Step 9. How many times did someone order more than one Canned Soda?

```python
# Nos quedamos con ventas de 'Canned Soda' agrupadas por cantidad vendida
ventas_canned = (chipo[chipo['item_name'] == 'Canned Soda']
                 .groupby('order_id')['quantity']
                 .sum())

# Buscamos los pedidos con más de 1 'Canned Soda'
pedidos_soda = (ventas_canned > 1).sum()
print('Hay', pedidos_soda, 'pedidos con más de 1 Canned Soda')
```
